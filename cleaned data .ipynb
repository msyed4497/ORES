```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv("measure of birth and death.csv")
df
```

```python
# Deleting columns
columns_to_remove=['CI_Min_Injury',	'CI_Max_Injury']
df=df.drop(columns=columns_to_remove)
df
```

```python
df.dtypes
```

```python
# Data Types Transformation
df['Min_Injury']=df['Min_Injury'].round(0).astype(int)
df['Max_Injury']=df['Max_Injury'].round(0).astype(int)
df['Total_Births']=df['Total_Births'].round(0).astype(int)
df
```

```python
df.dtypes
```

```python
# Replacing the negative values with NaN
df[df['Suicide']<0]=np.nan
df[df['CI_Min_Suicide']<0]=np.nan
df[df['CI_Max_Suicide']<0]=np.nan
df
```

```python
#  Handling NULL or Infinite Values
df=df.fillna('-')
df
```

```python
# Renaming the columns
df.rename(columns={'LBW':'Low_Birth_weight'},inplace=True)
df
```

```python
# Calculated field creation
df['Min_Injury'] = pd.to_numeric(df['Min_Injury'], errors='coerce')
df['Max_Injury'] = pd.to_numeric(df['Max_Injury'], errors='coerce')
df['Avg_Injury'] = (df['Min_Injury'] + df['Max_Injury']) / 2
df
```

```python
# DROPPING DUPLICATE ROWS:
df.drop_duplicates()
df
```
